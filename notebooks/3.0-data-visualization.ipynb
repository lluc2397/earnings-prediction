```python
ROOT_DIR = Path().resolve(strict=True).parent
DATA_DIR = f"{ROOT_DIR}/data/"

df = pd.read_csv(f"{DATA_DIR}/interim/new_features.csv", index_col=0)
short_df = pd.read_csv(f"{DATA_DIR}/interim/new_features_drops.csv", index_col=0)
```

```python
aapl_df = df[df["Ticker"] == "AAPL"]
intc_df = df[df["Ticker"] == "INTC"]
msft_df = df[df["Ticker"] == "MSFT"]
```

```python
aapl_df = aapl_df[(aapl_df["Fiscal Year"] >= 2008) & (aapl_df["Fiscal Year"] <= 2019)]
intc_df = intc_df[(intc_df["Fiscal Year"] >= 2008) & (intc_df["Fiscal Year"] <= 2019)]
msft_df = msft_df[(msft_df["Fiscal Year"] >= 2008) & (msft_df["Fiscal Year"] <= 2019)]
```

```python
# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.

x_ax = list(set(list(aapl_df["Fiscal Year"].values) + list(msft_df["Fiscal Year"].values)))

fig, ax = plt.subplots(layout='constrained')
ax.plot(aapl_df["Fiscal Year"], aapl_df["Revenue"], label="aapl")
ax.plot(intc_df["Fiscal Year"], intc_df["Revenue"], label="intc")
ax.plot(msft_df["Fiscal Year"], msft_df["Revenue"], label="msft")
ax.set_xlabel('Fiscal Year')  # Add an x-label to the axes.
ax.set_ylabel('Revenue')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend();  # Add a legend.
```

```python

width = 0.35  # the width of the bars

fig, ax = plt.subplots()

aapl_bar = ax.bar(aapl_df["Fiscal Year"], aapl_df["Revenue"], width, label="aapl")
intc_bar = ax.bar(intc_df["Fiscal Year"], intc_df["Revenue"], width, label="intc")
msft_bar = ax.bar(msft_df["Fiscal Year"], msft_df["Revenue"], width, label="msft")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x_ax, x_ax)
ax.legend()

ax.bar_label(aapl_bar, padding=3)
ax.bar_label(intc_bar, padding=3)
ax.bar_label(msft_bar, padding=3)

fig.tight_layout()

plt.show()
```
